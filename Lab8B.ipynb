```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer, OneHotEncoder
from sklearn.pipeline import Pipeline
```

```python
dataset=pd.read_csv('heart_2020_cleaned.csv')
```

```python
dataset=dataset.drop(['Asthma', 'DiffWalking'], axis=1)
```

Impute missing values in kidney disease using mode strategy using function transformer and in column transformer i use simpleimputer for sleep time using mean strategy and then convert heartdisease and kidney disease to numerical using one hot encoder

```python
#function transformer
def impute_KidneyDisease(X):
    X['KidneyDisease'] = X['KidneyDisease'].fillna(X['KidneyDisease'].mode()[0])  # Fill missing values
    print(X['KidneyDisease'])
    return X

# ColumnTransformer
preprocessor = ColumnTransformer(transformers=[
    ('SleepTime_imputer', SimpleImputer(strategy='mean'), ['SleepTime']),
    ('KidneyDisease_imputer', FunctionTransformer(impute_KidneyDisease), ['KidneyDisease']),
    ('ohe', OneHotEncoder(), ['HeartDisease','KidneyDisease'])
], remainder='passthrough')  # Pass through other columns like 'Fare'

# Transform the data
transformed_data = preprocessor.fit_transform(dataset)

# Output shape and transformed data
print("\nTransformed Data Shape:")
print(transformed_data.shape)
print(transformed_data)


transformed_df = pd.DataFrame(transformed_data)
print("\nTransformed DataFrame:")
print(transformed_df)
```
